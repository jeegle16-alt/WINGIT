```python
import sagemaker
from pipeline import get_pipeline

# SageMaker Studio 실행 역할
role = sagemaker.get_execution_role()

# 파이프라인 객체 생성
p = get_pipeline(
    pipeline_name="flight-price-xgb-pipeline",
    region="us-east-1",   # 명시 추천
    role=role,            # pipeline.py가 받도록 되어 있다면
)

# 파이프라인 정의 등록(Studio Pipelines에 표시됨)
p.upsert(role_arn=role)

# 실행
execution = p.start()
print("Execution ARN:", execution.arn)
```
