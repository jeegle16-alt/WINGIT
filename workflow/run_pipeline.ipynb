```python
import sagemaker
from sagemaker import get_execution_role

# 1) 파이프라인 가져오기
from pipeline import get_pipeline  # 파일 상단에 workflow/pipeline.py 라고 되어 있음 :contentReference[oaicite:3]{index=3}

region = sagemaker.Session().boto_region_name
role = get_execution_role()

pipeline = get_pipeline(
    region=region,
    role=role,
    pipeline_name="flight-price-xgb-pipeline",  # 원하는 이름
)

# 2) upsert: Studio Pipelines에 파이프라인 정의 등록/업데이트
pipeline.upsert(role_arn=role)

# 3) start: 실행 (파라미터 오버라이드 가능)
execution = pipeline.start(
    parameters={
        "EndpointName": "flight-price-xgb-endpoint",
        "ArtifactBucket": "sagemaker-studio-558853120313-37b94c10",
    }
)

print("Execution ARN:", execution.arn)
```
